```python
import numpy as np
import cv2
import json
```

```python
f = open("mask.json", "r")
data = json.load(f)
print(data)
```

```python
data = data["_via_img_metadata"]
print(data)
```

```python
img_dir = "../images/training_images"
mask_dir = "../images/mask"
```

```python
for key, value in data.items():
    filemane = value["filename"]
    img_path = f"{img_dir}/{filemane}"
    img = cv2.imread(img_path, cv2.IMREAD_COLOR)
    h, w, _ = img.shape
    
    mask = np.zeros((h, w))
    
    regions = value["regions"]
    for region in regions:
        shape = region["shape_attributes"]
        x_points = shape["all_points_x"]
        y_points = shape["all_points_y"]
        
        contour = []
        for x, y in zip(x_points, y_points):
            contour.append([x, y])
        contour = np.array(contour)
        
        cv2.drawContours(mask, [contour], -1, 255, -1)
        
    cv2.imwrite(f"{mask_dir}/{filemane}", mask)
```
